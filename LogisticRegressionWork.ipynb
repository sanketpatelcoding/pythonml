```python
import numpy as np
import pandas as pd
 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
pages = np.array([100, 150, 200, 250, 300, 350, 400, 450, 500]).reshape(-1, 1)
likes = np.array([0, 1, 1, 0, 1, 0, 0, 0, 0])
```

```python
model = LogisticRegression()
model.fit(pages, likes)
```

```python
book_pages = [[266]]
predict = model.predict(book_pages)
print(predict)
```

```python
plt.scatter(pages, likes, color='green')
plt.plot(pages, model.predict_proba(pages)[:, 1], color='red')
plt.show()
```

```python
df = pd.read_csv('heart.csv')
print(df.head())
```

```python
X = df[['chol', 'thalach']]
y = df['target']
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, shuffle=True, stratify=y)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
model = LogisticRegression(max_iter=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
print(y_predict[:5])
```

```python
acc_score = accuracy_score(y_test, y_predict)
print(acc_score)
```
