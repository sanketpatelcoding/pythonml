```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Load the CSV data
# Assuming the CSV is saved as 'home_data.csv' in your working directory
# If you're using the CSV string directly, you can use pd.read_csv(io.StringIO(csv_string))
data = pd.read_csv('home_data.csv')

# Prepare the data
# Create a binary target: 1 if price > median, 0 otherwise
median_price = data['price'].median()
data['price_class'] = (data['price'] > median_price).astype(int)

# Features (sqft_living) and target (price_class)
X = data[['sqft_living']].values  # Feature: square footage
y = data['price_class'].values    # Target: binary price class

# Fit logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Generate points for the decision boundary
x_range = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
y_prob = model.predict_proba(x_range)[:, 1]  # Probability of class 1

# Plot the data and decision boundary
plt.scatter(X, y, color='blue', label='Data points', alpha=0.5)
plt.plot(x_range, y_prob, color='red', label='Logistic curve')
plt.axhline(0.5, color='green', linestyle='--', label='Decision boundary (0.5)')
plt.xlabel('Square Footage (sqft_living)')
plt.ylabel('Probability (Price > Median)')
plt.title('Logistic Regression: Price Classification')
plt.legend()
plt.show()

# Print model accuracy
accuracy = model.score(X, y)
print(f"Model accuracy: {accuracy:.2f}")
```
