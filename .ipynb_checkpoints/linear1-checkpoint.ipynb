```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt


area = np.array([100, 165, 180, 200, 250, 300, 500]).reshape(-1, 1)
price1 = np.array([1, 2, 3, 7, 11, 30, 40])

model = LinearRegression()
model.fit(area, price1)


area_range = np.linspace(100, 500).reshape(-1, 1) 
predicted_prices = model.predict(area_range)


plt.scatter(area, price1, color='blue', label='real prices')   
plt.plot(area_range, predicted_prices, color='red')  
plt.xlabel('area')
plt.ylabel('price')

plt.show()
```
